# Advance House Price Prediction Feature Selection 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
```

```python
train_dataset = pd.read_csv('train.csv')
train_dataset.head()
```

###### We have already performed all the feature engineering in the dataset.

```python
#creating x_train and y_train
x_train = train_dataset.drop(labels=['SalePrice'], axis=1)
y_train = train_dataset[['SalePrice']]
```

## Applying Feature Selection
1. I specify the Lasso Regression model.
2. select a suitable alpha (equivalent of penalty),The bigger the alpha the less features that will be selected.
3. Then use the selectFromModel object from sklearn, which will select the features which coefficients are non-zero.

```python
#feature selection
feature_sel_model = SelectFromModel(Lasso(alpha=0.005, random_state=0))
feature_sel_model.fit(x_train, y_train)
```

Now if we execute <strong>feature_sel_model.get_support()</strong> it will give the array of True and False which is associated with independent feature or column. True means the corresponding feature is important and False means the corresponding feature is not important.

```python
feature_sel_model.get_support()
```

```python
#creating the list of important selected features w.r.t. dependent feature.
selected_features = x_train.columns[feature_sel_model.get_support()]
selected_features
```

```python
#features which are not selected
not_selected_features = [feature for feature in x_train.columns if feature not in selected_features]
```

```python
#cheking the numbers of important features and features which are not important w.r.t. dependent feature.

print(f'Total number of features :- {len(x_train.columns)}')
print(f'Number of selected features :- {len(selected_features)}')
print(f'Number of features which are not selected :- {len(not_selected_features)}')
```

```python
#important features
x_train = x_train[selected_features]
x_train.head()
```
